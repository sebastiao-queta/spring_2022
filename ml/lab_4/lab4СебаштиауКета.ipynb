### 7.Решите поставленную задачу регрессии в соответствии с заданием. При подборе параметров метода принятия решения используйте перекрёстную проверку (изучите возможные для изменения параметры метода регрессии). Вычислите коэффициент корреляции (Пирсона, Спирмена) между реальным значением и предсказаннымс учётом p-value

```python
# разделим полученные данные на данные для обучения и теста
X, y = df.drop(['price'], axis=1), df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 123)
```

### 7.1. Линейная регрессия без использования регуляризаторов

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
pred_regr = regr.predict(X_test)
print(pred_regr[:15] - y_test[:15])
```

```python
corr={}
corr['pearson'], _ = pearsonr(pred_regr,y_test)
corr['spearman'], _ = spearmanr(pred_regr,y_test)
print('Корреляция Пирсона: ', corr['pearson'])
print('Корреляция Спирмена: ', corr['spearman'])
```

### 7.2 Линейная регрессия, RIDGE регуляризатор

```python
reg = Ridge()
reg_grid = GridSearchCV(reg, {'alpha': np.arange(0, 3, 0.1)})
reg_grid.fit(X_train, y_train)
best_alpha = reg_grid.best_estimator_.alpha

reg = Ridge(alpha=best_alpha)
reg.fit(X_train, y_train)
reg_pred = reg.predict(X_test)

print(reg_pred[:15] - y_test[:15])
print(reg_pred)
```

```python
corr = {}
corr['pearson'], _ = pearsonr(reg_pred, y_test)
corr['spearman'], _ = spearmanr(reg_pred, y_test)
print('Корреляция Пирсона: ', corr['pearson'])
print('Корреляция Спирмена: ', corr['spearman'])
```
